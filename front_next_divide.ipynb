```python
import wrds
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import datetime
from scipy import stats
import pickle
from pandas.tseries.offsets import BDay
from pandas.tseries.offsets import DateOffset
```

```python
spx_opt = pd.read_csv('data\\spx_opt_monthly.csv')
spx_stock = pd.read_csv('data\\spx_stock.csv',header=None)
```

```python
spx_opt.date = pd.to_datetime(spx_opt.date)
spx_opt.exdate = pd.to_datetime(spx_opt.exdate)
spx_stock.loc[:,0] =  pd.to_datetime(spx_stock.loc[:,0])
```

```python
def get_series(tic_en,tic_opt):
    aapl_en,aapl_df = tic_en,tic_opt
    front_s,next_s,leap_s = pd.DataFrame(),pd.DataFrame(),pd.DataFrame()
    
    for i in range(len(aapl_en)):
        
        cur_date = aapl_en[i]
        print(cur_date)
        cur_df = aapl_df[aapl_df.date == cur_date]
        cur_exdates = pd.to_datetime(sorted(cur_df.exdate.unique()))

        cur_front = cur_df[cur_df.exdate == cur_exdates[0]]
        cur_next = cur_df[cur_df.exdate == cur_exdates[1]]

        front_s = pd.concat([front_s,cur_front])
        next_s = pd.concat([next_s,cur_next])
 
    return front_s,next_s
```

```python
spx_dates = pd.to_datetime(sorted(spx_opt.date.unique()))
spx_ft,spx_nx = get_series(spx_dates,spx_opt)
```

```python
spx_ft["mid_price"] = (spx_ft["best_offer"] +  spx_ft["best_bid"])/2
spx_nx["mid_price"] = (spx_nx["best_offer"] +  spx_nx["best_bid"])/2

spx_ft.to_csv('data\\spx_front.csv')
spx_nx.to_csv('data\\spx_next.csv')
```
